```python
import os
import shutil
from pathlib import Path

print("Starting Project Setup...")

# 1. Automatically find and copy the data from your Downloads folder!
home = str(Path.home())
download_path = os.path.join(home, 'Downloads', 'petrolgasprice.csv')
current_path = os.path.join(os.getcwd(), 'petrolgasprice.csv')

if os.path.exists(download_path):
    shutil.copy(download_path, current_path)
    print("✅ SUCCESS: Found petrolgasprice.csv in Downloads and copied it here!")
else:
    print(f"⚠️ WARNING: Could not find data at {download_path}. Please check the spelling in your Downloads folder.")

# 2. Create the folders
os.makedirs("src", exist_ok=True)
os.makedirs("tests", exist_ok=True)

# 3. Create standard files (Git, Environment, Init)
with open("src/__init__.py", "w") as f: f.write("")
with open("tests/__init__.py", "w") as f: f.write("")
with open(".gitignore", "w") as f: f.write("*.csv\n__pycache__/\n.ipynb_checkpoints/\n.pytest_cache/\n")
with open("environment.yml", "w") as f:
    f.write("name: petrol_resit_env\nchannels:\n  - defaults\n  - conda-forge\ndependencies:\n  - python=3.10\n  - pandas\n  - matplotlib\n  - seaborn\n  - statsmodels\n  - pytest\n  - jupyter\n")

# 4. Create Advanced Data Prep Script
with open("src/data_prep.py", "w") as f:
    f.write('''import pandas as pd

def load_data(filepath: str) -> pd.DataFrame:
    """Loads the CSV file into a pandas DataFrame."""
    return pd.read_csv(filepath)

def clean_data(df: pd.DataFrame) -> pd.DataFrame:
    """Cleans column names for modeling and removes missing data."""
    df_clean = df.rename(columns={
        'Inflation_Rate (%)': 'Inflation',
        'GDP_Growth (%)': 'GDP_Growth'
    })
    # COUNTER-INTUITIVE CHOICE: We drop missing values entirely instead of filling them with averages.
    # Why? Filling macro-economic gaps with averages distorts the variance of natural economic shocks.
    return df_clean.dropna()
''')

# 5. Create Advanced Visuals Script (Added Correlation Heatmap)
with open("src/visuals.py", "w") as f:
    f.write('''import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

def plot_crude_impact(df: pd.DataFrame) -> None:
    """Plots a scatter plot showing the relationship between crude oil and petrol prices."""
    plt.figure(figsize=(9, 5))
    sns.scatterplot(data=df, x='Crude_Oil_Price_USD_per_Barrel', y='Petrol_Price_USD_per_Liter', color='teal', alpha=0.6)
    plt.title("Impact of Global Crude Oil on Pump Prices")
    plt.xlabel("Crude Oil (USD / Barrel)")
    plt.ylabel("Petrol Price (USD / Liter)")
    plt.show()

def plot_correlation_heatmap(df: pd.DataFrame) -> None:
    """Plots a heatmap to visualize correlations between numeric macroeconomic variables."""
    numeric_df = df.select_dtypes(include=[np.number])
    plt.figure(figsize=(10, 8))
    sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    plt.title("Macroeconomic Correlation Heatmap")
    plt.tight_layout()
    plt.show()
''')

# 6. Create Advanced Modeling Script (Added Multicollinearity Check via VIF)
with open("src/modeling.py", "w") as f:
    f.write('''import pandas as pd
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

def run_regression(df: pd.DataFrame):
    """Runs an OLS regression to determine drivers of petrol prices."""
    formula = "Petrol_Price_USD_per_Liter ~ Crude_Oil_Price_USD_per_Barrel + Inflation + GDP_Growth"
    return smf.ols(formula, data=df).fit()

def check_multicollinearity(df: pd.DataFrame) -> pd.DataFrame:
    """Calculates Variance Inflation Factor (VIF) to check for feature correlation."""
    features = df[['Crude_Oil_Price_USD_per_Barrel', 'Inflation', 'GDP_Growth']].dropna()
    X = add_constant(features)
    vif_data = pd.DataFrame()
    vif_data["Feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
    return vif_data
''')

# 7. Create Tests
with open("tests/test_pipeline.py", "w") as f:
    f.write('''import pandas as pd
from src.data_prep import clean_data

def test_clean_data_unit():
    """Unit test: verifies column renaming and NA dropping."""
    df = pd.DataFrame({'Inflation_Rate (%)': [2.5, None], 'Petrol_Price_USD_per_Liter': [1.5, 1.6]})
    cleaned_df = clean_data(df)
    assert 'Inflation' in cleaned_df.columns
    assert len(cleaned_df) == 1
''')

print("✅ SUCCESS: Your complete advanced project architecture has been generated!")
```
